```python
dummy=pd.get_dummies(df['City'])
df=pd.concat([df,dummy],axis='columns')
df=df.drop('City',axis=1)
```

```python
def gen(x):
    if x=='Male':
        x= 0
    else:
        x= 1
    return x
df['Gender']=df['Gender'].apply(gen)
```

```python
def change(x):
    if x=='Yes':
        x=1
    else:
        x=0
    return x    
df['Illness']=df['Illness'].apply(change)
```

```python
#outline detection using z score
df['zscore']=(df.Income-df.Income.mean())/df.Income.std()
```
